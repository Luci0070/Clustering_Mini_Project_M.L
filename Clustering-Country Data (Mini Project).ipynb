# KMeans :

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
km=KMeans()
km.fit_transform(df1)
```

```python
elb=[]
sil=[]
cluster=range(2,9)
for r in range(2,9):
    km=KMeans(n_clusters=r)
    km.fit(df1[['PC1','PC2']])
    elb.append(km.inertia_)
    sil.append(silhouette_score(df1[['PC1','PC2']],km.labels_))
plt.plot(cluster,elb,marker='o') 
plt.xlabel('Cluster')
plt.ylabel('WCSS')
plt.title('Elbow Plot :')
plt.grid()
plt.show()
```

```python
clus=range(2,9)
plt.plot(cluster,sil,marker='o',color='lime')
plt.xlabel('Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Plot :')
plt.grid()
plt.show()
```

```python
km1=KMeans(n_clusters=3,random_state=0)
y_pre=km1.fit_predict(df1[['PC1','PC2']])
y_pre
```

```python
df1['ClusterID']=y_pre
```

```python
df11=df1[df1['ClusterID']==0]
df12=df1[df1['ClusterID']==1]
df13=df1[df1['ClusterID']==2]
plt.scatter(df11['PC1'],df11['PC2'],color='blue',label='cluster=0')
plt.scatter(df12['PC1'],df12['PC2'],color='black',label='cluster=1')
plt.scatter(df13['PC1'],df13['PC2'],color='green',label='cluster=2')
plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],color='red',marker='*')
plt.legend()
plt.show()
```

# Agglomerative Clustering :

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram
```

```python
ac=AgglomerativeClustering(n_clusters=None,distance_threshold=0)
ac.fit_predict(df1[['PC1','PC2']])
```

```python
plt.figure(figsize=(45,30))
L=hierarchy.linkage(ac.children_)
D=dendrogram(L,p=13,truncate_mode='level',color_threshold=20)
```

```python
sc=[]
for x in range(2,6):
    ac1=AgglomerativeClustering(n_clusters=x)
    ac1.fit(df1[['PC1','PC2']])
    sc.append(silhouette_score(df1[['PC1','PC2']],ac1.labels_))
    print('For ',x,' Cluster. The score is = '+str(silhouette_score(df1[['PC1','PC2']],ac1.labels_)))
```

```python
clus1=range(2,6)
plt.plot(clus1,sc,marker='o',color='lime')
plt.xlabel('Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Plot :')
plt.grid()
plt.show()
```

```python
agc=AgglomerativeClustering(n_clusters=3)
y_pre_agc=agc.fit_predict(df1[['PC1','PC2']])
y_pre_agc
```

```python
df1['ClusterID']=y_pre_agc
```

```python
df1a=df1[df1['ClusterID']==0]
df1b=df1[df1['ClusterID']==1]
df1c=df1[df1['ClusterID']==2]
plt.scatter(df1a['PC1'],df1a['PC2'],color='blue',label='cluster=0')
plt.scatter(df1b['PC1'],df1b['PC2'],color='black',label='cluster=1')
plt.scatter(df1c['PC1'],df1c['PC2'],color='green',label='cluster=2')
plt.legend()
plt.show()
```

# DBSCAN :

```python
from sklearn.cluster import DBSCAN
```

```python
eps=np.linspace(0.6,1.5,10)
sc1=[]
for y in eps:
    db1=DBSCAN(eps=y)
    db1.fit(df1[['PC1','PC2']])
    sc1.append(silhouette_score(df1[['PC1','PC2']],db1.labels_))
    print('For eps = ',y,' Silhouette score is = '+str(silhouette_score(df1[['PC1','PC2']],db1.labels_)))
```

```python
eps=np.linspace(0.5,1.5,10)
plt.plot(eps,sc1,color='lime',marker='o')
plt.xlabel('Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Plot :')
plt.grid()
plt.show()
```

```python
db=DBSCAN(eps=1.4,min_samples=10)
y_pre_db=db.fit_predict(df1[['PC1','PC2']])
y_pre_db
```

```python
df1['ClusterID']=y_pre_db
```

```python
df1a1=df1[df1['ClusterID']==0]
df1b1=df1[df1['ClusterID']==-1]
plt.scatter(df1a1['PC1'],df1a1['PC2'],color='black',label='Cluster')
plt.scatter(df1b1['PC1'],df1b1['PC2'],color='red',marker='*',label='Outlier')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
